# Stepwise Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor as DTR
from sklearn.feature_selection import f_regression as fitlm


spy = pd.read_csv('./CSV/SPY2004_2014.csv')
```

```python
spy['Date'] = pd.to_datetime(spy['Date'])
spy.set_index('Date', inplace=True)
plt.figure(figsize=(20,10))
plt.plot(spy.Close, color ='green')
plt.show()
```

# Test and train data set split

Now, we will split data into training and test data set. 

1. First 50% of data is used for training and remaining data for testing.
2. X_train and y_train are training dataset.
3. X_test and y_test are test dataset.

```python
spy['ret1'] = spy.Close.pct_change(1) #1 Day lookback
spy['ret2'] = spy.Close.pct_change(2) #2 Day lookback
spy['ret5'] = spy.Close.pct_change(5) #5 Day lookback
spy['ret20'] = spy.Close.pct_change(20) #20 Day lookback

y = spy.Close.pct_change(1).shift(-1)
spy = spy.dropna()
spy = spy.drop(['Open', 'High', 'Low', 'Adj Close', 'Volume', 'Close'], axis=1)

spy.head()
```

```python
split_percentage = 0.5
split = int(split_percentage*len(spy))
X = spy.copy()
# Train data set
X_train = X[:split]
y_train = y[:split] 

# Test data set
X_test = X[split:]
y_test = y[split:]
X_train.head()
```

```python
tree_reg = DTR(min_samples_leaf=100)
tree_reg.fit(X_train, y_train)
```

```python
import graphviz
from sklearn.tree import export_graphviz

from IPython.display import display

display(graphviz.Source(export_graphviz(tree_reg)))
```

# See performance of predictor

Calculate CAGR(Compound annual growth rate), Sharpe Ratio, Calmar Ratio
The split comes from calculating the path (in the Decision Tree) to get to the largest positive future return and the largest negative future return.

```python
spy['positions'] = 0
#The following split is from the tree in the book
spy['positions'] = np.where((spy.ret2 >= 0.01531), -1, spy.positions )
spy['positions'] = np.where(((spy.ret2 < 0.01531) & (spy.ret1 < -0.01392)), 1, spy.positions)
```

# Performance on actual SPY

```python
spy = spy[1188:]

spy['DailyReturn'] = spy.ret1.shift(-1) * spy.positions
spy['cumret'] =np.cumprod(spy['DailyReturn'] + 1) -1
spy.cumret.plot(figsize=(20,5))

plt.ylabel("Strategy Returns (%)")
plt.show()
```

# Performance on Chan's SPY

```python
spyChan = pd.read_csv('./CSV/SPY_Chan.csv')
spyChan = spyChan[1188:]
spyChan['positions'] = 0
#The following split is from the tree in the book
spyChan['positions'] = np.where((spyChan.ret2 >= 0.01531), -1, spyChan.positions )
spyChan['positions'] = np.where(((spyChan.ret2 < 0.01531) & (spyChan.ret1 < -0.01392)), 1, spyChan.positions)

spyChan['DailyReturn'] = spyChan.ret1.shift(-1) * spyChan.positions
spyChan['cumret'] =np.cumprod(spyChan['DailyReturn'] + 1) -1
spyChan.cumret.plot(figsize=(20,5))

plt.ylabel("Strategy Returns (%)")
plt.show()
```

```python
CAGR = (1+spyChan.dropna().tail(1).cumret.item())**(252/1179)-1
Sharpe = (252**.5) * (spyChan.DailyReturn.mean() / spyChan.DailyReturn.std())

print ('CAGR: ', CAGR)
print ('Sharpe: ', Sharpe)
```
