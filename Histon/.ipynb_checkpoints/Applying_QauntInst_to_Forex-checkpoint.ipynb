```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

euus_df = pd.read_csv ('./CSV/EURUSD2014_2018.csv')
```

```python
euus_df.head()
```

### Begin Selecting Data to Plot

```python
#Include only first n number of rows for better visibility
euus = euus_df.iloc[0:99]
euus = euus [['date', 'close']]

euus.set_index('date', inplace=True)

euus.index = pd.to_datetime(euus.index)
```

```python
%matplotlib inline 
#Assures plot will appear in jupyter notebook

plt.plot(euus)

plt.show ()
```

### Plot Better

```python
# Plot Size
plt.figure(figsize = (14, 5)) 

#Line Color
plt.plot(euus, 'r')

# bo: Blue circles at every discrete data point.
#plt.plot(euus,'bo')


#Grid
plt.grid(True)

#Axis and Plot Titles
plt.title('EUR/USD Closing Price')
plt.xlabel('Trading Dates')
plt.ylabel('EURUSD Closing Price')

#Show Plot
plt.show()
```

### Repeat Above, But with Open and Close Price

```python
euus2 = euus_df [['date', 'open', 'close']]

euus2.set_index('date', inplace=True)

euus2.index = pd.to_datetime(euus2.index)
```

```python
# Plot Size
plt.figure(figsize = (20, 8)) 

#Line Color

plt.plot(euus2["close"], lw=1.5, label = 'Close Price')
plt.plot(euus2["open"], lw=1.5, label = 'Open Price')


#Grid
plt.grid(True)

plt.legend(loc=0)

#This helps us tighten the figure margins
plt.axis ('tight')


#Axis and Plot Titles
plt.title('EUR/USD Closing Price')
plt.xlabel('Dates')
plt.ylabel('EURUSD Index')

#Show Plot
plt.show()
```

### Scatter Plots

```python
euus2 = euus_df.iloc[0:999]
euus2 = euus2 [['date', 'open','close']]

plt.figure (figsize = (20,15))

#plt.axis([1.374,1.378,1.374,1.378])
plt.scatter(euus2["close"], euus2["open"], marker='o')

plt.grid(True)
plt.xlabel ('Close')
plt.ylabel ('Open')
plt.title('Scatter Plot')
plt.show() 
```
