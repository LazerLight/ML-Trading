### DataFrame.diff()

This method returns the 'difference' between the current observation and the previous observation.

### DataFrame.pct_change()

This method returns the percentage change of the current observation with the previous observation.

### DataFrame.var()
Variance

### DataFrame.std()
Standard deviation

### DataFrame.cov()
Covariance

### DataFrame.corr()
Correlation

### DataFrame.kurt()
Kurtosis


```python
print('\n==================\n Diff Method \n==================\n', 
      df['Open'].diff().head())

print('\n==================\n Diff Method \n==================\n', 
      df['Close'].diff().head())

print('\n==================\n Diff Method \n==================\n', 
      df['Close'].diff().var().head())

print('\n==================\n Diff Method \n==================\n', 
      df['Close'].diff().std().head())

print('\n==================\n Diff Method \n==================\n', 
      df['Close'].diff().std())
```

## Indexing using .loc()

```python
df.shape #How many rows and columns

df.loc[:,['Close', 'Open']] #Select only all rows of Close and Open 

df.loc[:4,['Close', 'Open']] #Select only the first 5 rows of Close and Open 

df.loc[:4] #Select only the first 5 rows of all the columns
```

## Indexing using .iloc()

Another way to perform indexing is using the 'iloc()' method.

iloc is extremely useful when your data is not labelled and you need to refer to columns using their integer location instead


```python
df.iloc[:4] #Select the first 4 rows

df.iloc[:4, 1:3] # Select the first 4 rows and columns from 2 to 3
```

## Missing Values

Missing values are values that are absent from the data frame. Usually, all the data frames that you would work on, would be large and there will be a case of 'missing values' in most of them. <br>
<br>
Hence, it becomes important for you to learn how to handle these missing values.

```python
df.isnull() #Returns boolean for null values for each data point

df['Close'].fillna(method='backfill') #Backfill any null values
df.fillna(method='ffill') #Forward fill for entire df

df.dropna() # By default, dropna() will exclude or drop all the rows which has even one NaN value in it

```
